````python
#1) Seu código aqui

def consultar_dados_tesouro(uf):
    # URL da API
    url = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
    
    # Parâmetros da consulta
    params = {'uf': uf}

    # Realiza a consulta
    r = requests.get(url, params=params)

    # Verifica se a requisição foi bem-sucedida
    if r.status_code == 200:
        # Converte a resposta para formato JSON
        data_json = r.json()
        
        # Verifica se há itens na resposta
        if 'items' in data_json:
            # Cria um DataFrame com base nos itens
            df = pd.DataFrame(data_json['items'])
            
            # Filtra o DataFrame com base nos parâmetros fornecidos
            df_final = df[(df['uf'] == uf)]
            
            return df_final
        else:
            print("Nenhum item encontrado na resposta.")
    else:
        print(f"Falha na requisição. Código de status: {r.status_code}")
        return none
````
