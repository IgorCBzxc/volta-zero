#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

2.1 Faça uma média móvel em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

````python
df_ma = df['Adj Close'].rolling(window=90).mean()
df_ma
````

2.2 Utilizando a fução rolling(), calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

````python
df_std = df['Adj Close'].rolling(window=90).std()
df_std
````

2.3 Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
df_ma_adjclose = df_ma.stack().reset_index()
df_ma_adjclose.columns = ['data', 'ativo', 'valor']

fig = px.line(data_frame=df_ma_adjclose,
              x='data',
              y='valor',
              color='ativo',
              template='plotly_white')
fig.show()
````

2.4 Monte um gráfico de linhas para os desvios padrão em janela móvel

````python
df_std_adjclose = df_std.stack().reset_index()
df_std_adjclose.columns = ['data', 'ativo', 'valor']

fig = px.line(data_frame=df_std_adjclose,
              x='data',
              y='valor',
              color='ativo',
              template='plotly_white')
fig.show()
````

2.5 Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.


Comparando os gráficos de média móvel e desvio padrão das séries temporais do Bitcoin USD (BTC-USD) desde 2020 com o gráfico anterior que mostrava a volatilidade, notamos um evento significativo entre o início de 2021 e metade de 2022. Essa variação anômala é evidente nos dois gráficos, sendo mais precisa no desvio padrão e menos na média móvel nesse período. Ambos os gráficos, no entanto, apontam para uma alta volatilidade, como visto no gráfico anterior com indicadores sem janela móvel. Concluímos que há semelhanças nas análises, indicando a necessidade de investigar as causas por trás dessas variações anômalas durante as anomalias identificadas.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

3.1 Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base

````python
df_last60 = df.xs(key='BTC-USD',
                  level=1,
                  axis=1).tail(n=60)

print(f'Quantidade de linhas do período selecionado: {df_last60.shape[0]}.\n')
df_last60
````

3.2 Monte um gráfico de candlestick para esses dados.

````python
layout = Layout(paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='white',
                height=600)

graph = {
    'x': df_last60.index,
    'open': df_last60.Open,
    'close': df_last60.Close,
    'high': df_last60.High,
    'low': df_last60.Low,
    'type': 'candlestick',
    'name': 'BTC-USD',
    'showlegend': True
}

grafico = graph_objects.Figure(data=[graph],
                               layout_title='Índice Bitcoin',
                               layout=layout)

grafico.update_xaxes(showgrid=True,
                     gridwidth=1,
                     gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True,
                     gridwidth=1,
                     gridcolor='LightGrey')

grafico
````
