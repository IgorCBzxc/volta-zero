````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from scipy.stats import ttest_ind

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

# Sexo

````python
sns.set(style="whitegrid")

# Criação do gráfico de barras
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='data_ref', hue='sexo', data=df, palette='pastel')

# Adiciona título ao gráfico
plt.title('Distribuição de sexo ao Longo do Tempo')

# Adiciona legenda e rotaciona em 45 graus
ax.legend(title='Sexo', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

# Exibe o gráfico
plt.show()
````
