1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from scipy.stats import ttest_ind

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# Divida o DataFrame em dois grupos com base na posse de imóvel
possui_imovel = df[df['posse_de_imovel'] == 1]['renda']
nao_possui_imovel = df[df['posse_de_imovel'] == 0]['renda']

# Execute o teste t independente
t_stat, p_value = ttest_ind(possui_imovel, nao_possui_imovel)

# Imprima o resultado
print(f'Teste t para Posse de Imóvel:\nEstatística t = {t_stat}\nValor-p = {p_value}')
````

O valor-p é menor que 0.05 (um nível comum de significância), indicando que há evidências estatísticas para rejeitar a hipótese nula. Portanto, podemos concluir que a posse de imóvel parece discriminar a renda de forma estatisticamente significativa.

````python
# Divida o DataFrame em dois grupos com base na posse de veículo
possui_veiculo = df[df['posse_de_veiculo'] == 1]['renda']
nao_possui_veiculo = df[df['posse_de_veiculo'] == 0]['renda']

# Execute o teste t independente
t_stat, p_value = ttest_ind(possui_veiculo, nao_possui_veiculo)

# Imprima o resultado
print(f'Teste t para Posse de Veículo:\nEstatística t = {t_stat}\nValor-p = {p_value}')
````

Neste caso, o valor-p é extremamente baixo (praticamente zero), o que indica uma forte evidência estatística contra a hipótese nula. Portanto, podemos concluir que a posse de veículo também discrimina a renda de forma estatisticamente significativa.

Em resumo, com base nos resultados dos testes t, tanto a posse de imóvel quanto a posse de veículo parecem ser boas preditoras da renda, indicando que há diferenças estatisticamente significativas nas médias de renda entre os grupos que possuem e não possuem esses bens.
