Uma instituição financeira quer conhecer melhor o perfil de renda de seus novos clientes para diversos fins, por exemplo, melhor dimensionar o limite de cartões de crédito dos novos clientes, sem necessariamente solicitar olerites ou documentações que impactem na experiência do seu cliente.

Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

As variáveis são intuitivas - note que há uma variável 'index' que é um identificador do cliente e que em geral o ```read_csv``` do pandas coloca também uma variável sequencial.

Estes dados estão no arquivo ```previsao_de_renda.csv```. Carregue-o em um *dataframe*.

1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('previsao_de_renda.csv')

df.head()
````

2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
df.corr()
````

````python
# As variáveis tempo_emprego e posse_de_veiculo são as que possuem maior correlação

sns.set(style="whitegrid")  # Define o estilo do gráfico

# Criação do gráfico de dispersão
plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Scatter plot usando seaborn
sns.scatterplot(x='renda', y='tempo_emprego', hue='posse_de_veiculo', data=df, palette='viridis')

# Adiciona legendas aos eixos
plt.xlabel('Renda')
plt.ylabel('Tempo de Emprego')

# Adiciona título ao gráfico
plt.title('Gráfico de Dispersão com Três Variáveis')

# Exibe a legenda
plt.legend(title='Posse de Veículo')

# Exibe o gráfico
plt.show()
````

3) Na sua opinião, há outliers na variável renda?

````python
# Define o estilo do gráfico
sns.set(style="whitegrid")

# Criação do box plot
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
sns.boxplot(x='renda', data=df, orient='v', width=0.5)

# Adiciona título ao gráfico
plt.title('Box Plot da Variável Renda')

# Exibe o gráfico
plt.show()
````

Analisando o boxplot gerado, acredito que a variável renda tenha outliers sim.

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
# Calcula o logaritmo da variável 'renda'
df['log_renda'] = np.log1p(df['renda'])  # Usando np.log1p para evitar problemas com valores zero

sns.set(style="whitegrid")  # Define o estilo do gráfico

# Criação do gráfico de dispersão
plt.figure(figsize=(10, 6))  # Define o tamanho da figura

# Scatter plot usando seaborn
sns.scatterplot(x='log_renda', y='tempo_emprego', hue='posse_de_veiculo', data=df, palette='viridis')

# Adiciona legendas aos eixos
plt.xlabel('Log Renda')
plt.ylabel('Tempo de Emprego')

# Adiciona título ao gráfico
plt.title('Gráfico de Dispersão com Três Variáveis')

# Exibe a legenda
plt.legend(title='Posse de Veículo')

# Exibe o gráfico
plt.show()
````

````python
# Define o estilo do gráfico
sns.set(style="whitegrid")

# Criação do box plot para o logaritmo da renda
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
sns.boxplot(x='log_renda', data=df, orient='v', width=0.5)

# Adiciona título ao gráfico
plt.title('Box Plot do Logaritmo da Variável Renda')

# Exibe o gráfico
plt.show()
````

5) A aplicação do LOG você acha que melhorou ou piorou a análise?

Após gerar os gráficos acima e aplicar o logaritmo à variável de renda, percebo uma melhoria na análise. O gráfico de dispersão agora apresenta uma definição mais clara, e a amplitude dos outliers parece ter diminuído. O uso do logaritmo pode ser benéfico ao lidar com distribuições assimétricas, proporcionando uma visão mais equilibrada dos dados e facilitando a identificação de padrões.
