#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Define os intervalos e rótulos para categorizar a variável APGAR5
intervalos = [0, 3, 5, 7, 10]
rotulos = ['asfixia grave', 'asfixia moderada', 'asfixia leve', 'normal']

# Categoriza a variável APGAR5
df['APGAR5_CATEGORIA'] = pd.cut(df['APGAR5'], bins=intervalos, labels=rotulos, include_lowest=True)

# Visualiza o DataFrame resultante
print(df[['APGAR5', 'APGAR5_CATEGORIA']])
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='GESTACAO', y='APGAR5', hue='SEXO', data=df, palette='muted', split=True)
plt.title('Distribuição de APGAR5 por Sexo e Tempo de Gestação')
plt.xlabel('Tempo de Gestação')
plt.ylabel('APGAR5')
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Criação da tabela cruzada
tabela_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5'])

# Visualização da tabela cruzada
tabela_cruzada.head()
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Criação da tabela cruzada com percentuais de linha
tabela_cruzada_percent = pd.crosstab(df['GESTACAO'], df['APGAR5'], normalize='index') * 100

# Visualização da tabela cruzada com percentuais de linha
tabela_cruzada_percent.head()
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Converta a coluna 'GESTACAO' para o tipo numérico
df['GESTACAO'] = pd.to_numeric(df['GESTACAO'], errors='coerce')

# Filtre o DataFrame para gestações com menos de 37 semanas
df_prematuro = df[df['GESTACAO'] < 37]

# Certifique-se de que a coluna 'APGAR5' é do tipo numérico
df_prematuro['APGAR5'] = pd.to_numeric(df_prematuro['APGAR5'], errors='coerce')

# Certifique-se de que há pelo menos uma linha e uma coluna na tabela cruzada
if not df_prematuro.empty and len(df_prematuro['APGAR5'].unique()) > 1:
    # Criação da tabela cruzada com percentuais de linha
    tabela_cruzada_percent = pd.crosstab(df_prematuro['GESTACAO'], df_prematuro['APGAR5'], normalize='index') * 100

    # Gráfico de barras
    plt.figure(figsize=(12, 8))
    sns.barplot(x=tabela_cruzada_percent.columns, y=tabela_cruzada_percent.iloc[:, 0], color='skyblue')
    plt.title('Distribuição de APGAR5 para Gestações com Menos de 37 Semanas')
    plt.xlabel('APGAR5')
    plt.ylabel('Percentual de Linha')
    plt.show()
else:
    print("Não há dados suficientes para criar o gráfico.")
```
