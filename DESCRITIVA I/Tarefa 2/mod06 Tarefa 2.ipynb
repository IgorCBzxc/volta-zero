```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

```python
df.columns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Crie o boxplot da idade da mãe
plt.figure(figsize=(10, 6))
sns.boxplot(x='IDADEMAE', data=df, color='skyblue')
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Crie o boxplot do peso do bebê
plt.figure(figsize=(10, 6))
sns.boxplot(x='PESO', data=df, color='skyblue')
plt.title('Boxplot do peso do bebe')
plt.xlabel('Gramas')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Histograma da idade da mãe

plt.figure(figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins=30, kde=True, color='skyblue')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Histograma do peso do bebê

plt.figure(figsize=(10, 6))
sns.histplot(df['PESO'], bins=30, kde=True, color='skyblue')
plt.title('Histograma do peso do bebê')
plt.xlabel('Gramas')
plt.ylabel('Frequência')
plt.show()
```
