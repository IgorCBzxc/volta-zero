#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Gráfico de barras para APGAR1
plt.figure(figsize=(8, 5))
df['APGAR1'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')
plt.show()

# Gráfico de barras para APGAR5
plt.figure(figsize=(8, 5))
df['APGAR5'].value_counts().sort_index().plot(kind='bar', color='lightcoral')
plt.title('Distribuição de APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Contagem')
plt.show()

# Gráfico de barras para SEXO
plt.figure(figsize=(6, 4))
df['SEXO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribuição de SEXO')
plt.xlabel('SEXO')
plt.ylabel('Contagem')
plt.show()

# Gráfico de barras para GESTACAO
plt.figure(figsize=(6, 4))
df['GESTACAO'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Distribuição de GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Contagem')
plt.show()

# Gráfico de barras para ESTCIVMAE
plt.figure(figsize=(8, 5))
df['ESTCIVMAE'].value_counts().plot(kind='bar', color='gold')
plt.title('Distribuição de ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Contagem')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Gráfico de barras para APGAR5 médio por CONSULTAS
plt.figure(figsize=(10, 6))
df.groupby('CONSULTAS')['APGAR5'].mean().sort_values().plot(kind='bar', color='lightcoral')
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Criação de um painel com 2 linhas e 2 colunas
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico de barras para SEXO
sns.countplot(x='SEXO', data=df, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Distribuição de SEXO')

# Gráfico de barras para GESTACAO com rótulos rotacionados em 45 graus
sns.countplot(x='GESTACAO', data=df, ax=axes[0, 1], palette='magma')
axes[0, 1].set_title('Distribuição de GESTACAO')
axes[0, 1].tick_params(axis='x', rotation=45)  # Rotaciona os rótulos do eixo x

# Gráfico de barras para ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=df, ax=axes[1, 0], palette='plasma')
axes[1, 0].set_title('Distribuição de ESTCIVMAE')

# Gráfico de barras para CONSULTAS
sns.countplot(x='CONSULTAS', data=df, ax=axes[1, 1], palette='inferno')
axes[1, 1].set_title('Distribuição de CONSULTAS')

# Ajusta o layout
plt.tight_layout()

# Mostra o painel
plt.show()
```
